# 0.初始化

**描述**：

初始化函数库的引用，包括typing, sys（将当前目录加入系统路径）...

定义基本类：
1. 单向链表


````python
from typing import List, Tuple, Dict, Optional
import sys
sys.path.append("./")

INT_MAX = 2**31 - 1
INT_MIN = -2**31


class ListNode:
    # Definition for singly-linked list.
    def __init__(self, val: int = 0, next=None):
        self.val = val
        self.next = next

    def _len(self):
        length = 1
        cur = self.next
        while cur:
            length += 1
            cur = cur.next
        return length

    def __len__(self):
        return self._len()

    @classmethod
    def build(cls, l: List[int]) -> "ListNode":
        head = cls(l[0])
        cur = head
        for i in l[1:]:
            cur.next = cls(i)
            cur = cur.next
        return head

    @classmethod
    def equal(cls, l1: "ListNode", l2: "ListNode"):
        if len(l1) != len(l2):
            return False
        else:
            while l1 and l2:
                if l1.val == l2.val:
                    l1 = l1.next
                    l2 = l2.next
                else:
                    return False
            return True
````

# 1.两数之和

**描述**：

给定一个整数数组 nums 和一个整数目标值 target，请你在该数组中找出和为目标值 target 的那两个整数，并返回它们的数组下标。

你可以假设每种输入只会对应一个答案。但是，数组中同一个元素在答案里不能重复出现。

你可以按任意顺序返回答案。

**示例 1**：

```python
输入：nums = [2,7,11,15], target = 9
输出：[0,1]
解释：因为 nums[0] + nums[1] == 9 ，返回 [0, 1]。
```

**示例 2**：

```python
输入：nums = [3,2,4], target = 6
输出：[1,2]
```

**示例 3**：

```python
输入：nums = [3,3], target = 6
输出：[0,1]
```


````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["Traverse", "inList", "dictHash", "dictHash2", "dictHash3"]


class Solution(object):
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        Solute = getattr(self, SOLUTE[-2])
        rtn = Solute(nums, target)
        return rtn

    def Traverse(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        for i in range(n-1):
            for j in range(i+1, n):
                sumNum = nums[i] + nums[j]
                if sumNum == target:
                    return [i, j]

    def inList(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        for i in range(n-1):
            resid = target - nums[i]
            if resid in nums[i+1:]:
                j = i + 1 + nums[i+1:].index(resid)
                return [i, j]

    def dictHash(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        hashMap = dict(zip(nums, list(range(n))))
        for i in range(n):
            resid = target - nums[i]
            j = hashMap.get(resid)
            if j and j != i:
                return [i, j]

    def dictHash2(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        hashMap = dict()
        idx_list = []
        for i in range(n):
            resid = target - nums[i]
            if resid in hashMap:
                return [hashMap[resid], i]
            else:
                hashMap[nums[i]] = i

    def dictHash3(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        hashMap = dict()
        idx_list = []
        for i in range(n):
            resid = target - nums[i]
            if resid in hashMap:
                idx_list.append([hashMap[resid], i])
            else:
                hashMap[nums[i]] = i
        return sorted(idx_list, key=lambda x: x[0])[0]


test_data = [
    {
        "nums": [2, 7, 11, 15],
        "target": 9,
        "rtn": [0, 1]
    },
    {
        "nums": [3, 2, 4],
        "target": 6,
        "rtn": [1, 2]
    },
    {
        "nums": [3, 3],
        "target": 6,
        "rtn": [0, 1]
    }]

s = Solution()
tester.uniTest(s.twoSum, test_data)
````

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["Traverse", ]


class Solution(object):
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        Solute = getattr(self, SOLUTE[0])
        rtn = Solute(l1, l2)
        return rtn

    def Traverse(self, l1: ListNode, l2: ListNode) -> ListNode:
        sumHead = ListNode()
        cur = sumHead
        remain = 0
        while l1 or l2:
            num1, num2 = 0, 0
            if l1:
                num1 = l1.val
                l1 = l1.next
            if l2:
                num2 = l2.val
                l2 = l2.next
            sumNum = (num1 + num2 + remain) % 10
            cur.next = ListNode(sumNum)
            cur = cur.next
            remain = (num1 + num2 + remain) // 10
        if remain != 0:
            cur.next = ListNode(remain)
        return sumHead.next


test_data = [
    {
        "l1": ListNode.build([2, 4, 3]),
        "l2": ListNode.build([5, 6, 4]),
        "rtn": ListNode.build([7, 0, 8])
    },
    {
        "l1": ListNode.build([0]),
        "l2": ListNode.build([0]),
        "rtn": ListNode.build([0])
    },
    {
        "l1": ListNode.build([9, 9, 9, 9, 9, 9, 9]),
        "l2": ListNode.build([9, 9, 9, 9]),
        "rtn": ListNode.build([8, 9, 9, 9, 0, 0, 0, 1])
    }]

s = Solution()
tester.uniTest(s.addTwoNumbers, test_data, ListNode.equal)
````

# 3.无重复子串

**描述**：

给定一个字符串 s ，请你找出其中不含有重复字符的 **最长子串** 的长度。

**示例 1**：

```python
输入: s = "abcabcbb"
输出: 3 
解释: 因为无重复字符的最长子串是 "abc"，所以其长度为 3。
```

**示例 2**：

```python
输入: s = "bbbbb"
输出: 1
解释: 因为无重复字符的最长子串是 "b"，所以其长度为 1。
```

**示例 3**：

```python
输入: s = "pwwkew"
输出: 3
解释: 因为无重复字符的最长子串是"wke"，所以其长度为 3。
    请注意，你的答案必须是 子串 的长度，"pwke"是一个子序列，不是子串。
```

**示例 4**：

```python
输入: s = ""
输出: 0
```

**提示**：

+ 0 <= s.length <= $ 5 * 10^{4} $
+ s 由英文字母、数字、符号和空格组成

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["DP1", "DP2", "DP3", "dictHash"]


class Solution(object):
    def lengthOfLongestSubstring(self, s: str) -> int:
        Solute = getattr(self, SOLUTE[1])
        rtn = Solute(s)
        return rtn

    def DP1(self, s: str) -> int:
        n = len(s)
        if n == 0:
            return 0
        maxLen = 1
        for i in range(n-1):
            tmpLen = 1
            for j in range(i+1, n):
                if s[j] in s[i:j]:
                    break
                else:
                    tmpLen += 1
            # maxLen = max(maxLen, tmpLen)
            maxLen = maxLen if maxLen >= tmpLen else tmpLen
        return maxLen
    
    def DP2(self, s: str) -> int:
        n = len(s)
        if n == 0:
            return 0
        maxLen = 1
        for i in range(n-1):
            tmpLen = 1
            if n - i <= maxLen:
                break
            for j in range(i+1, n):
                if s[j] in s[i:j]:
                    break
                else:
                    tmpLen += 1
            maxLen = maxLen if maxLen >= tmpLen else tmpLen
        return maxLen

    def DP3(self, s: str) -> int:
        n = len(s)
        if n == 0:
            return 0
        maxLen = 1
        start = 0
        end = 1
        while end < n:
            while end < n and s[end] not in s[start:end]:
                end += 1
            maxLen = max(maxLen, end-start)
            if end != n:
                start += s[start:end].index(s[end]) + 1
        return maxLen

    def dictHash(self, s: str) -> int:
        n = len(s)
        if n == 0:
            return 0
        hashMap = dict()
        maxLen = 0
        i = -1  # i是左指针
        for j in range(n):  # j是右指针
            if s[j] in hashMap:  # 若当前元素在之前出现过，更新左指针
                # 当之前出现的元素在左右指针中间，左指针更新为之前元素下标，若不在中间，左指针不变
                i = max(i, hashMap[s[j]])
            hashMap[s[j]] = j  # 将当前元素加入哈希表中
            maxLen = max(maxLen, j - i)
        return maxLen


test_data = [
    {
        "s": "abcabcbb",
        "rtn": 3
    },
    {
        "s": "bbbbb",
        "rtn": 1
    },
    {
        "s": "pwwkew",
        "rtn": 3
    },
    {
        "s": "",
        "rtn": 0
    }]

s = Solution()
tester.uniTest(s.lengthOfLongestSubstring, test_data)
````

# 4.寻找两个正序数组的中位数

**描述**：

给定两个大小分别为 m 和 n 的正序（从小到大）数组 nums1 和 nums2。请你找出并返回这两个正序数组的 中位数 。

算法的时间复杂度应该为 O(log (m+n)) 。

**示例 1**：

```python
输入：nums1 = [1,3], nums2 = [2]
输出：2.00000
解释：合并数组 = [1,2,3] ，中位数 2
```

**示例 2**：

```python
输入：nums1 = [1,2], nums2 = [3,4]
输出：2.50000
解释：合并数组 = [1,2,3,4] ，中位数 (2 + 3) / 2 = 2.5
```

**示例 3**：

```python
输入：nums1 = [0,0], nums2 = [0,0]
输出：0.00000
```

**示例 4**：

```python
输入：nums1 = [], nums2 = [1]
输出：1.00000
```

**示例 5**：

```python
输入：nums1 = [2], nums2 = []
输出：2.00000
```

**提示**：

+ nums1.length == m
+ nums2.length == n
+ 0 <= m <= 1000
+ 0 <= n <= 1000
+ 1 <= m + n <= 2000
+ $ -10^6 $ <= nums1[i], nums2[i] <= $ 10^6 $

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["sortedList", "DP1", ]


class Solution(object):
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        Solute = getattr(self, SOLUTE[0])
        rtn = Solute(nums1, nums2)
        return rtn

    def sortedList(self, nums1: List[int], nums2: List[int]) -> float:
        num = sorted(nums1 + nums2)
        n = len(num)
        odd = n % 2
        if odd:
            return float(num[n // 2])
        return (num[n // 2] + num[n // 2 - 1]) / 2

    def DP1(self, nums1: List[int], nums2: List[int]) -> float:
        m = len(nums1)
        n = len(nums2)
        i, j = 0, 0
        conlist = []
        while True:
            if i < m and j < n:
                if nums1[i] < nums2[j]:
                    conlist.append(nums1[i])
                    i += 1
                else:
                    conlist.append(nums2[j])
                    j += 1
            elif i >= m and j < n:
                conlist.append(nums2[j])
                j += 1
            elif i < m and j >= n:
                conlist.append(nums1[i])
                i += 1
            else:
                break
        if (m + n) % 2 != 0:
            return conlist[(m + n) // 2]
        else:
            return (conlist[(m + n) // 2 - 1] + conlist[(m + n) // 2]) / 2


test_data = [
    {
        "nums1": [1, 3],
        "nums2": [2],
        "rtn": 2.00000
    },
    {
        "nums1": [1, 2],
        "nums2": [3, 4],
        "rtn": 2.50000
    },
    {
        "nums1": [0, 0],
        "nums2": [0, 0],
        "rtn": 0.00000
    },
    {
        "nums1": [],
        "nums2": [1],
        "rtn": 1.00000
    },
    {
        "nums1": [2],
        "nums2": [],
        "rtn": 2.00000
    }]

s = Solution()
tester.uniTest(s.findMedianSortedArrays, test_data)
````

# 5.最长回文子串

**描述**：

给你一个字符串 s，找到 s 中最长的回文子串。

**示例 1**：

```python
输入：s = "babad"
输出："bab"
解释："aba" 同样是符合题意的答案。
```

**示例 2**：

```python
输入：s = "cbbd"
输出："bb"
```

**示例 3**：

```python
输入：s = "a"
输出："a"
```

**示例 4**：

```python
输入：s = "ac"
输出："a"
```

**提示**：

+ 1 <= s.length <= 1000
+ s 仅由数字和英文字母（大写和/或小写）组成

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["DP1", "DP2", "expandCenter"]


class Solution(object):
    def longestPalindrome(self, s: str) -> str:
        Solute = getattr(self, SOLUTE[1])
        rtn = Solute(s)
        return rtn

    def DP1(self, s: str) -> str:
        n = len(s)
        dp = [[False for _ in range(n)] for _ in range(n)]
        maxLen = 0
        longestSeq = ""
        for j in range(n):
            for i in range(0, j + 1):
                if j - i <= 1:
                    if s[i] == s[j]:
                        dp[i][j] = True
                        if maxLen < j - i + 1:
                            maxLen = j - i + 1
                            longestSeq = s[i:j + 1]
                else:
                    if s[i] == s[j] and dp[i + 1][j - 1]:
                        dp[i][j] = True
                        if maxLen < j - i + 1:
                            maxLen = j - i + 1
                            longestSeq = s[i:j + 1]
        return longestSeq

    def DP2(self, s: str) -> str:
        n = len(s)
        if n <= 1:
            return '' if n == 0 else s[0]
        else:
            def isPalindrome(ss):
                m = len(ss)
                odd = m % 2
                left = m // 2
                right = m // 2
                if odd:
                    right += 1
                return ss[:left] == ss[right:][::-1]

            # dp[i]代表以下标i结尾的最长回文子串长度
            dp = [1 for _ in range(n)]
            for i in range(1, n):
                for j in range(i - 1 - dp[i - 1], i):
                    # print(dp)
                    if j < 0:
                        continue
                    if isPalindrome(s[j:i + 1]):
                        dp[i] = i + 1 - j
                        break
        maxLen = max(dp)
        maxIndex = dp.index(maxLen)
        return s[maxIndex - maxLen + 1:maxIndex + 1]

    def expandCenter(self, s: str) -> str:
        def expand(left, right):
            while left >= 0 and right < len(s) and s[left] == s[right]:
                left -= 1
                right += 1
            return left + 1, right - 1

        start, end = 0, 0
        for i in range(len(s)):
            left1, right1 = expand(i, i)
            left2, right2 = expand(i, i + 1)
            if right1 - left1 > end - start:
                start, end = left1, right1
            if right2 - left2 > end - start:
                start, end = left2, right2
        return s[start: end + 1]


test_data = [
    {
        "s": "babad",
        "rtn": "bab"
    },
    {
        "s": "cbbd",
        "rtn": "bb"
    },
    {
        "s": "a",
        "rtn": "a"
    },
    {
        "s": "ac",
        "rtn": "a"
    }]

s = Solution()
tester.uniTest(s.longestPalindrome, test_data)
````

# 6.Z字形变换

**描述**：

将一个给定字符串 s 根据给定的行数 numRows ，以从上往下、从左到右进行 Z 字形排列。

比如输入字符串为 "PAYPALISHIRING" 行数为 3 时，排列如下：

```python
P   A   H   N
A P L S I I G
Y   I   R
```

之后，你的输出需要从左往右逐行读取，产生出一个新的字符串，比如："PAHNAPLSIIGYIR"。

请你实现这个将字符串进行指定行数变换的函数：

```python
string convert(string s, int numRows);
```

**示例 1**：

```python
输入：s = "PAYPALISHIRING", numRows = 3
输出："PAHNAPLSIIGYIR"
```

**示例 2**：

```python
输入：s = "PAYPALISHIRING", numRows = 4
输出："PINALSIGYAHRPI"
解释：
P     I    N
A   L S  I G
Y A   H R
P     I
```

**示例 3**：

```python
输入：s = "A", numRows = 1
输出："A"
```

**提示**：

+ 1 <= s.length <= 1000
+ s 由英文字母（小写和大写）、',' 和 '.' 组成
+ 1 <= numRows <= 1000

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["TraverseClassify", "ForwardBack", ]


class Solution(object):
    def convert(self, s: str, numRows: int) -> str:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(s, numRows)
        return rtn

    def TraverseClassify(self, s: str, numRows: int) -> str:
        n = len(s)
        if numRows == 1 or numRows >= n:
            return s
        result = ""
        N = 2 * numRows - 2
        pairs = n // N
        remain = n % N
        if remain == 0:
            for i in range(numRows):
                for j in range(pairs):
                    result += s[j * N + i]
                    if i != 0 and i != numRows - 1:
                        result += s[j * N + N - i]
        elif remain <= numRows:
            for i in range(numRows):
                for j in range(pairs + 1):
                    if j < pairs:
                        result += s[j * N + i]
                    if i <= remain - 1 and j == pairs:
                        result += s[j * N + i]
                    if i != 0 and i != numRows - 1 and j < pairs:
                        result += s[j * N + N - i]
        else:
            for i in range(numRows):
                for j in range(pairs + 1):
                    result += s[j * N + i]
                    if i != 0 and i != numRows - 1 and j < pairs:
                        result += s[j * N + N - i]
                    if i >= N - remain + 1 and i != numRows - 1 and j == pairs:
                        result += s[j * N + N - i]
        return result

    def ForwardBack(self, s: str, numRows: int) -> str:
        if numRows < 2:
            return s
        result = ["" for _ in range(numRows)]
        i, direction = 0, -1
        for ch in s:
            result[i] += ch
            if i == 0 or i == numRows - 1:
                direction = -direction
            i += direction
        return "".join(result)


test_data = [
    {
        "s": "PAYPALISHIRING",
        "numRows": 3,
        "rtn": "PAHNAPLSIIGYIR"
    },
    {
        "s": "PAYPALISHIRING",
        "numRows": 4,
        "rtn": "PINALSIGYAHRPI"
    },
    {
        "s": "A",
        "numRows": 1,
        "rtn": "A"
    }]

s = Solution()
tester.uniTest(s.convert, test_data)
````

# 7.整数反转

**描述**：

给你一个 32 位的有符号整数 x ，返回将 x 中的数字部分反转后的结果。

如果反转后整数超过 32 位的有符号整数的范围 $ [−2^{31}, 2^{31} − 1] $ ，就返回 0。

**假设环境不允许存储 64 位整数（有符号或无符号）。**

**示例 1**：

```python
输入：x = 123
输出：321
```

**示例 2**：

```python
输入：x = -123
输出：-321
```

**示例 3**：

```python
输入：x = 120
输出：21
```

**示例 4**：

```python
输入：x = 0
输出：0
```

**提示**：

+ $ -2^{31} \le x \le 2^{31} - 1 $

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["Traverse", "strReverse", "TraverseSum"]


class Solution(object):
    def reverse(self, x: int) -> int:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(x)
        return rtn

    def Traverse(self, x: int) -> int:
        neg = True if x < 0 else False
        if x == 0:
            return x
        posx = abs(x)
        nums = []
        flag = False
        while posx != 0:
            remain = posx % 10
            if remain != 0 or flag:
                nums.append(remain)
                flag = True
            posx = posx // 10
        n = len(nums)
        res = 0
        for i in range(n):
            res += nums[i] * 10**(n - 1 - i)
        if res > 2**31 - 1 or res < - 2**31:
            return 0
        if neg:
            return -res
        return res

    def strReverse(self, x: int) -> int:
        y = (-1 if x < 0 else 1) * int(str(abs(x))[::-1])
        return y if y.bit_length() < 32 else 0

    def TraverseSum(self, x: int) -> int:
        y, res = abs(x), 0
        # 则其数值范围为 [−2^31,  2^31 − 1]
        boundry = (1 << 31) - 1 if x > 0 else 1 << 31
        while y != 0:
            res = res * 10 + y % 10
            if res > boundry:
                return 0
            y //= 10
        return res if x > 0 else -res


test_data = [
    {
        "x": 123,
        "rtn": 321
    },
    {
        "x": 123,
        "rtn": 321
    },
    {
        "x": 123,
        "rtn": 321
    },
    {
        "x": 123,
        "rtn": 321
    }]

s = Solution()
tester.uniTest(s.reverse, test_data)
````

**描述**：

请你来实现一个 myAtoi(string s) 函数，使其能将字符串转换成一个 32 位有符号整数（类似 C/C++ 中的 atoi 函数）。

函数 myAtoi(string s) 的算法如下：

+ 读入字符串并丢弃无用的前导空格
+ 检查下一个字符（假设还未到字符末尾）为正还是负号，读取该字符（如果有）。 确定最终结果是负数还是正数。 如果两者都不存在，则假定结果为正。
+ 读入下一个字符，直到到达下一个非数字字符或到达输入的结尾。字符串的其余部分将被忽略。
+ 将前面步骤读入的这些数字转换为整数（即，"123" -> 123， "0032" -> 32）。如果没有读入数字，则整数为 0 。必要时更改符号（从步骤 2 开始）。
+ 如果整数数超过 32 位有符号整数范围 [$−2^{31}$, $2^{31} − 1$] ，需要截断这个整数，使其保持在这个范围内。具体来说，小于 $−2^{31}$ 的整数应该被固定为 $−2^{31}$ ，大于 $2^{31} − 1$ 的整数应该被固定为 $2^{31} − 1$ 。
+ 返回整数作为最终结果。

**注意**：

+ 本题中的空白字符只包括空格字符 ' ' 。
+ 除前导空格或数字后的其余字符串外，**请勿忽略**任何其他字符。

**示例 1**：

```python
输入：s = "42"
输出：42
解释：加粗的字符串为已经读入的字符，插入符号是当前读取的字符。
第 1 步："42"（当前没有读入字符，因为没有前导空格）
         ^
第 2 步："42"（当前没有读入字符，因为这里不存在 '-' 或者 '+'）
         ^
第 3 步："42"（读入 "42"）
           ^
解析得到整数 42 。
由于 "42" 在范围 [-2^{31}, 2^{31} - 1] 内，最终结果为 42 。
```

**示例 2**：

```python
第 1 步："   -42"（读入前导空格，但忽视掉）
            ^
第 2 步："   -42"（读入 '-' 字符，所以结果应该是负数）
             ^
第 3 步："   -42"（读入 "42"）
               ^
解析得到整数 -42 。
由于 "-42" 在范围 [-2^{31}, 2^{31} - 1] 内，最终结果为 -42 。
```

```python
输入：s = "-91283472332"

          ^
第 3 步："-91283472332"（读入 "91283472332"）
                     ^
解析得到整数 -91283472332 。
由于 -91283472332 小于范围 [-2^{31}, 2^{31} - 1] 的下界，最终结果被截断为 -2^{31} = -2147483648 。
```

**提示**：

+ 0 <= s.length <= 200
+ s 由英文字母（大写和小写）、数字（0-9）、' '、'+'、'-' 和 '.' 组成

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["Traverse", "stateTransition"]


class Solution(object):
    def myAtoi(self, s: str) -> int:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(s)
        return rtn

    def Traverse(self, s: str) -> int:
        res = 0
        neg = False
        if not s:
            return res
        while s and s[0] == ' ':
            s = s[1:]
        if not s:
            return res
        if s[0] == '-':
            neg = True
            s = s[1:]
        elif s[0] == '+':
            s = s[1:]
        i = 0
        while i < len(s) and s[i] in "0123456789":
            i += 1
        s = s[:i]
        if not s:
            return res
        while s and s[0] == '0':
            s = s[1:]
        if not s:
            return res
        while s:
            res = res * 10 + int(s[0])
            s = s[1:]
        if neg and -res <= -2**31:
            return -2**31
        if not neg and res >= 2**31 - 1:
            return 2**31 - 1
        return -res if neg else res

    def stateTransition(self, s: str) -> int:
        states = [
            {' ': 0, 'd': 2, 's': 1, 'x': 3},  # 0
            {'d': 2, ' ': 3, 's': 3, 'x': 3},  # 1
            {'d': 2, ' ': 3, 's': 3, 'x': 3},  # 2
            {'d': 3, ' ': 3, 's': 3, 'x': 3},  # 3
        ]

        flag, ans, cur = 1, 0, 0
        for i in s:
            if i == ' ':
                t = ' '
            elif i in '+-':
                t = 's'
            elif i.isdigit():
                t = 'd'
            else:
                t = 'x'
            cur = states[cur][t]
            if cur == 2:
                ans = ans * 10 + int(i)
                ans = min(INT_MAX, ans) if flag == 1 else min(-INT_MIN, ans)
            elif cur == 1:
                flag = 1 if i == '+' else -1
        return flag * ans


test_data = [
    {
        "s": "42",
        "rtn": 42
    },
    {
        "s": "   -42",
        "rtn": -42
    },
    {
        "s": "4193 with words",
        "rtn": 4193
    },
    {
        "s": "words and 987",
        "rtn": 0
    },
    {
        "s": "-91283472332",
        "rtn": -2147483648
    },
    {
        "s": "00000-42a1234",
        "rtn": 0
    },
    {
        "s": "-2147483648",
        "rtn": -2147483648
    }]

s = Solution()
tester.uniTest(s.myAtoi, test_data)
````
