# 0.初始化

**描述**：

初始化函数库的引用，包括typing, sys（将当前目录加入系统路径）...

定义基本类：
1. 单向链表


````python
from typing import List, Tuple, Dict, Optional
import sys
sys.path.append("./")

INT_MAX = 2**31 - 1
INT_MIN = -2**31


class ListNode:
    # Definition for singly-linked list.
    def __init__(self, val: int = 0, next=None):
        self.val = val
        self.next = next

    def _len(self):
        length = 1
        cur = self.next
        while cur:
            length += 1
            cur = cur.next
        return length

    def __len__(self):
        return self._len()

    @classmethod
    def build(cls, l: List[int]) -> "ListNode":
        head = cls(l[0])
        cur = head
        for i in l[1:]:
            cur.next = cls(i)
            cur = cur.next
        return head

    @classmethod
    def equal(cls, l1: "ListNode", l2: "ListNode"):
        if len(l1) != len(l2):
            return False
        else:
            while l1 and l2:
                if l1.val == l2.val:
                    l1 = l1.next
                    l2 = l2.next
                else:
                    return False
            return True
````

# 1.两数之和

**描述**：

给定一个整数数组 nums 和一个整数目标值 target，请你在该数组中找出和为目标值 target 的那两个整数，并返回它们的数组下标。

你可以假设每种输入只会对应一个答案。但是，数组中同一个元素在答案里不能重复出现。

你可以按任意顺序返回答案。

**示例 1**：

```python
输入：nums = [2,7,11,15], target = 9
输出：[0,1]
解释：因为 nums[0] + nums[1] == 9 ，返回 [0, 1]。
```

**示例 2**：

```python
输入：nums = [3,2,4], target = 6
输出：[1,2]
```

**示例 3**：

```python
输入：nums = [3,3], target = 6
输出：[0,1]
```


````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["Traverse", "inList", "dictHash", "dictHash2", "dictHash3"]


class Solution(object):
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        Solute = getattr(self, SOLUTE[-2])
        rtn = Solute(nums, target)
        return rtn

    def Traverse(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        for i in range(n-1):
            for j in range(i+1, n):
                sumNum = nums[i] + nums[j]
                if sumNum == target:
                    return [i, j]

    def inList(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        for i in range(n-1):
            resid = target - nums[i]
            if resid in nums[i+1:]:
                j = i + 1 + nums[i+1:].index(resid)
                return [i, j]

    def dictHash(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        hashMap = dict(zip(nums, list(range(n))))
        for i in range(n):
            resid = target - nums[i]
            j = hashMap.get(resid)
            if j and j != i:
                return [i, j]

    def dictHash2(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        hashMap = dict()
        idx_list = []
        for i in range(n):
            resid = target - nums[i]
            if resid in hashMap:
                return [hashMap[resid], i]
            else:
                hashMap[nums[i]] = i

    def dictHash3(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        hashMap = dict()
        idx_list = []
        for i in range(n):
            resid = target - nums[i]
            if resid in hashMap:
                idx_list.append([hashMap[resid], i])
            else:
                hashMap[nums[i]] = i
        return sorted(idx_list, key=lambda x: x[0])[0]


test_data = [
    {
        "nums": [2, 7, 11, 15],
        "target": 9,
        "rtn": [0, 1]
    },
    {
        "nums": [3, 2, 4],
        "target": 6,
        "rtn": [1, 2]
    },
    {
        "nums": [3, 3],
        "target": 6,
        "rtn": [0, 1]
    }]

s = Solution()
tester.uniTest(s.twoSum, test_data)
````

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["Traverse", ]


class Solution(object):
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        Solute = getattr(self, SOLUTE[0])
        rtn = Solute(l1, l2)
        return rtn

    def Traverse(self, l1: ListNode, l2: ListNode) -> ListNode:
        sumHead = ListNode()
        cur = sumHead
        remain = 0
        while l1 or l2:
            num1, num2 = 0, 0
            if l1:
                num1 = l1.val
                l1 = l1.next
            if l2:
                num2 = l2.val
                l2 = l2.next
            sumNum = (num1 + num2 + remain) % 10
            cur.next = ListNode(sumNum)
            cur = cur.next
            remain = (num1 + num2 + remain) // 10
        if remain != 0:
            cur.next = ListNode(remain)
        return sumHead.next


test_data = [
    {
        "l1": ListNode.build([2, 4, 3]),
        "l2": ListNode.build([5, 6, 4]),
        "rtn": ListNode.build([7, 0, 8])
    },
    {
        "l1": ListNode.build([0]),
        "l2": ListNode.build([0]),
        "rtn": ListNode.build([0])
    },
    {
        "l1": ListNode.build([9, 9, 9, 9, 9, 9, 9]),
        "l2": ListNode.build([9, 9, 9, 9]),
        "rtn": ListNode.build([8, 9, 9, 9, 0, 0, 0, 1])
    }]

s = Solution()
tester.uniTest(s.addTwoNumbers, test_data, ListNode.equal)
````

# 3.无重复子串

**描述**：

给定一个字符串 s ，请你找出其中不含有重复字符的 **最长子串** 的长度。

**示例 1**：

```python
输入: s = "abcabcbb"
输出: 3 
解释: 因为无重复字符的最长子串是 "abc"，所以其长度为 3。
```

**示例 2**：

```python
输入: s = "bbbbb"
输出: 1
解释: 因为无重复字符的最长子串是 "b"，所以其长度为 1。
```

**示例 3**：

```python
输入: s = "pwwkew"
输出: 3
解释: 因为无重复字符的最长子串是"wke"，所以其长度为 3。
    请注意，你的答案必须是 子串 的长度，"pwke"是一个子序列，不是子串。
```

**示例 4**：

```python
输入: s = ""
输出: 0
```

**提示**：

+ 0 <= s.length <= $ 5 * 10^{4} $
+ s 由英文字母、数字、符号和空格组成

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["DP1", "DP2", "DP3", "dictHash"]


class Solution(object):
    def lengthOfLongestSubstring(self, s: str) -> int:
        Solute = getattr(self, SOLUTE[1])
        rtn = Solute(s)
        return rtn

    def DP1(self, s: str) -> int:
        n = len(s)
        if n == 0:
            return 0
        maxLen = 1
        for i in range(n-1):
            tmpLen = 1
            for j in range(i+1, n):
                if s[j] in s[i:j]:
                    break
                else:
                    tmpLen += 1
            # maxLen = max(maxLen, tmpLen)
            maxLen = maxLen if maxLen >= tmpLen else tmpLen
        return maxLen
    
    def DP2(self, s: str) -> int:
        n = len(s)
        if n == 0:
            return 0
        maxLen = 1
        for i in range(n-1):
            tmpLen = 1
            if n - i <= maxLen:
                break
            for j in range(i+1, n):
                if s[j] in s[i:j]:
                    break
                else:
                    tmpLen += 1
            maxLen = maxLen if maxLen >= tmpLen else tmpLen
        return maxLen

    def DP3(self, s: str) -> int:
        n = len(s)
        if n == 0:
            return 0
        maxLen = 1
        start = 0
        end = 1
        while end < n:
            while end < n and s[end] not in s[start:end]:
                end += 1
            maxLen = max(maxLen, end-start)
            if end != n:
                start += s[start:end].index(s[end]) + 1
        return maxLen

    def dictHash(self, s: str) -> int:
        n = len(s)
        if n == 0:
            return 0
        hashMap = dict()
        maxLen = 0
        i = -1  # i是左指针
        for j in range(n):  # j是右指针
            if s[j] in hashMap:  # 若当前元素在之前出现过，更新左指针
                # 当之前出现的元素在左右指针中间，左指针更新为之前元素下标，若不在中间，左指针不变
                i = max(i, hashMap[s[j]])
            hashMap[s[j]] = j  # 将当前元素加入哈希表中
            maxLen = max(maxLen, j - i)
        return maxLen


test_data = [
    {
        "s": "abcabcbb",
        "rtn": 3
    },
    {
        "s": "bbbbb",
        "rtn": 1
    },
    {
        "s": "pwwkew",
        "rtn": 3
    },
    {
        "s": "",
        "rtn": 0
    }]

s = Solution()
tester.uniTest(s.lengthOfLongestSubstring, test_data)
````

# 4.寻找两个正序数组的中位数

**描述**：

给定两个大小分别为 m 和 n 的正序（从小到大）数组 nums1 和 nums2。请你找出并返回这两个正序数组的 中位数 。

算法的时间复杂度应该为 O(log (m+n)) 。

**示例 1**：

```python
输入：nums1 = [1,3], nums2 = [2]
输出：2.00000
解释：合并数组 = [1,2,3] ，中位数 2
```

**示例 2**：

```python
输入：nums1 = [1,2], nums2 = [3,4]
输出：2.50000
解释：合并数组 = [1,2,3,4] ，中位数 (2 + 3) / 2 = 2.5
```

**示例 3**：

```python
输入：nums1 = [0,0], nums2 = [0,0]
输出：0.00000
```

**示例 4**：

```python
输入：nums1 = [], nums2 = [1]
输出：1.00000
```

**示例 5**：

```python
输入：nums1 = [2], nums2 = []
输出：2.00000
```

**提示**：

+ nums1.length == m
+ nums2.length == n
+ 0 <= m <= 1000
+ 0 <= n <= 1000
+ 1 <= m + n <= 2000
+ $ -10^6 $ <= nums1[i], nums2[i] <= $ 10^6 $

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["sortedList", "DP1", ]


class Solution(object):
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        Solute = getattr(self, SOLUTE[0])
        rtn = Solute(nums1, nums2)
        return rtn

    def sortedList(self, nums1: List[int], nums2: List[int]) -> float:
        num = sorted(nums1 + nums2)
        n = len(num)
        odd = n % 2
        if odd:
            return float(num[n // 2])
        return (num[n // 2] + num[n // 2 - 1]) / 2

    def DP1(self, nums1: List[int], nums2: List[int]) -> float:
        m = len(nums1)
        n = len(nums2)
        i, j = 0, 0
        conlist = []
        while True:
            if i < m and j < n:
                if nums1[i] < nums2[j]:
                    conlist.append(nums1[i])
                    i += 1
                else:
                    conlist.append(nums2[j])
                    j += 1
            elif i >= m and j < n:
                conlist.append(nums2[j])
                j += 1
            elif i < m and j >= n:
                conlist.append(nums1[i])
                i += 1
            else:
                break
        if (m + n) % 2 != 0:
            return conlist[(m + n) // 2]
        else:
            return (conlist[(m + n) // 2 - 1] + conlist[(m + n) // 2]) / 2


test_data = [
    {
        "nums1": [1, 3],
        "nums2": [2],
        "rtn": 2.00000
    },
    {
        "nums1": [1, 2],
        "nums2": [3, 4],
        "rtn": 2.50000
    },
    {
        "nums1": [0, 0],
        "nums2": [0, 0],
        "rtn": 0.00000
    },
    {
        "nums1": [],
        "nums2": [1],
        "rtn": 1.00000
    },
    {
        "nums1": [2],
        "nums2": [],
        "rtn": 2.00000
    }]

s = Solution()
tester.uniTest(s.findMedianSortedArrays, test_data)
````

# 5.最长回文子串

**描述**：

给你一个字符串 s，找到 s 中最长的回文子串。

**示例 1**：

```python
输入：s = "babad"
输出："bab"
解释："aba" 同样是符合题意的答案。
```

**示例 2**：

```python
输入：s = "cbbd"
输出："bb"
```

**示例 3**：

```python
输入：s = "a"
输出："a"
```

**示例 4**：

```python
输入：s = "ac"
输出："a"
```

**提示**：

+ 1 <= s.length <= 1000
+ s 仅由数字和英文字母（大写和/或小写）组成

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["DP1", "DP2", "expandCenter"]


class Solution(object):
    def longestPalindrome(self, s: str) -> str:
        Solute = getattr(self, SOLUTE[1])
        rtn = Solute(s)
        return rtn

    def DP1(self, s: str) -> str:
        n = len(s)
        dp = [[False for _ in range(n)] for _ in range(n)]
        maxLen = 0
        longestSeq = ""
        for j in range(n):
            for i in range(0, j + 1):
                if j - i <= 1:
                    if s[i] == s[j]:
                        dp[i][j] = True
                        if maxLen < j - i + 1:
                            maxLen = j - i + 1
                            longestSeq = s[i:j + 1]
                else:
                    if s[i] == s[j] and dp[i + 1][j - 1]:
                        dp[i][j] = True
                        if maxLen < j - i + 1:
                            maxLen = j - i + 1
                            longestSeq = s[i:j + 1]
        return longestSeq

    def DP2(self, s: str) -> str:
        n = len(s)
        if n <= 1:
            return '' if n == 0 else s[0]
        else:
            def isPalindrome(ss):
                m = len(ss)
                odd = m % 2
                left = m // 2
                right = m // 2
                if odd:
                    right += 1
                return ss[:left] == ss[right:][::-1]

            # dp[i]代表以下标i结尾的最长回文子串长度
            dp = [1 for _ in range(n)]
            for i in range(1, n):
                for j in range(i - 1 - dp[i - 1], i):
                    # print(dp)
                    if j < 0:
                        continue
                    if isPalindrome(s[j:i + 1]):
                        dp[i] = i + 1 - j
                        break
        maxLen = max(dp)
        maxIndex = dp.index(maxLen)
        return s[maxIndex - maxLen + 1:maxIndex + 1]

    def expandCenter(self, s: str) -> str:
        def expand(left, right):
            while left >= 0 and right < len(s) and s[left] == s[right]:
                left -= 1
                right += 1
            return left + 1, right - 1

        start, end = 0, 0
        for i in range(len(s)):
            left1, right1 = expand(i, i)
            left2, right2 = expand(i, i + 1)
            if right1 - left1 > end - start:
                start, end = left1, right1
            if right2 - left2 > end - start:
                start, end = left2, right2
        return s[start: end + 1]


test_data = [
    {
        "s": "babad",
        "rtn": "bab"
    },
    {
        "s": "cbbd",
        "rtn": "bb"
    },
    {
        "s": "a",
        "rtn": "a"
    },
    {
        "s": "ac",
        "rtn": "a"
    }]

s = Solution()
tester.uniTest(s.longestPalindrome, test_data)
````

# 6.Z字形变换

**描述**：

将一个给定字符串 s 根据给定的行数 numRows ，以从上往下、从左到右进行 Z 字形排列。

比如输入字符串为 "PAYPALISHIRING" 行数为 3 时，排列如下：

```python
P   A   H   N
A P L S I I G
Y   I   R
```

之后，你的输出需要从左往右逐行读取，产生出一个新的字符串，比如："PAHNAPLSIIGYIR"。

请你实现这个将字符串进行指定行数变换的函数：

```python
string convert(string s, int numRows);
```

**示例 1**：

```python
输入：s = "PAYPALISHIRING", numRows = 3
输出："PAHNAPLSIIGYIR"
```

**示例 2**：

```python
输入：s = "PAYPALISHIRING", numRows = 4
输出："PINALSIGYAHRPI"
解释：
P     I    N
A   L S  I G
Y A   H R
P     I
```

**示例 3**：

```python
输入：s = "A", numRows = 1
输出："A"
```

**提示**：

+ 1 <= s.length <= 1000
+ s 由英文字母（小写和大写）、',' 和 '.' 组成
+ 1 <= numRows <= 1000

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["TraverseClassify", "ForwardBack", ]


class Solution(object):
    def convert(self, s: str, numRows: int) -> str:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(s, numRows)
        return rtn

    def TraverseClassify(self, s: str, numRows: int) -> str:
        n = len(s)
        if numRows == 1 or numRows >= n:
            return s
        result = ""
        N = 2 * numRows - 2
        pairs = n // N
        remain = n % N
        if remain == 0:
            for i in range(numRows):
                for j in range(pairs):
                    result += s[j * N + i]
                    if i != 0 and i != numRows - 1:
                        result += s[j * N + N - i]
        elif remain <= numRows:
            for i in range(numRows):
                for j in range(pairs + 1):
                    if j < pairs:
                        result += s[j * N + i]
                    if i <= remain - 1 and j == pairs:
                        result += s[j * N + i]
                    if i != 0 and i != numRows - 1 and j < pairs:
                        result += s[j * N + N - i]
        else:
            for i in range(numRows):
                for j in range(pairs + 1):
                    result += s[j * N + i]
                    if i != 0 and i != numRows - 1 and j < pairs:
                        result += s[j * N + N - i]
                    if i >= N - remain + 1 and i != numRows - 1 and j == pairs:
                        result += s[j * N + N - i]
        return result

    def ForwardBack(self, s: str, numRows: int) -> str:
        if numRows < 2:
            return s
        result = ["" for _ in range(numRows)]
        i, direction = 0, -1
        for ch in s:
            result[i] += ch
            if i == 0 or i == numRows - 1:
                direction = -direction
            i += direction
        return "".join(result)


test_data = [
    {
        "s": "PAYPALISHIRING",
        "numRows": 3,
        "rtn": "PAHNAPLSIIGYIR"
    },
    {
        "s": "PAYPALISHIRING",
        "numRows": 4,
        "rtn": "PINALSIGYAHRPI"
    },
    {
        "s": "A",
        "numRows": 1,
        "rtn": "A"
    }]

s = Solution()
tester.uniTest(s.convert, test_data)
````

# 7.整数反转

**描述**：

给你一个 32 位的有符号整数 x ，返回将 x 中的数字部分反转后的结果。

如果反转后整数超过 32 位的有符号整数的范围 $ [−2^{31}, 2^{31} − 1] $ ，就返回 0。

**假设环境不允许存储 64 位整数（有符号或无符号）。**

**示例 1**：

```python
输入：x = 123
输出：321
```

**示例 2**：

```python
输入：x = -123
输出：-321
```

**示例 3**：

```python
输入：x = 120
输出：21
```

**示例 4**：

```python
输入：x = 0
输出：0
```

**提示**：

+ $ -2^{31} \le x \le 2^{31} - 1 $

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["Traverse", "strReverse", "TraverseSum"]


class Solution(object):
    def reverse(self, x: int) -> int:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(x)
        return rtn

    def Traverse(self, x: int) -> int:
        neg = True if x < 0 else False
        if x == 0:
            return x
        posx = abs(x)
        nums = []
        flag = False
        while posx != 0:
            remain = posx % 10
            if remain != 0 or flag:
                nums.append(remain)
                flag = True
            posx = posx // 10
        n = len(nums)
        res = 0
        for i in range(n):
            res += nums[i] * 10**(n - 1 - i)
        if res > 2**31 - 1 or res < - 2**31:
            return 0
        if neg:
            return -res
        return res

    def strReverse(self, x: int) -> int:
        y = (-1 if x < 0 else 1) * int(str(abs(x))[::-1])
        return y if y.bit_length() < 32 else 0

    def TraverseSum(self, x: int) -> int:
        y, res = abs(x), 0
        # 则其数值范围为 [−2^31,  2^31 − 1]
        boundry = (1 << 31) - 1 if x > 0 else 1 << 31
        while y != 0:
            res = res * 10 + y % 10
            if res > boundry:
                return 0
            y //= 10
        return res if x > 0 else -res


test_data = [
    {
        "x": 123,
        "rtn": 321
    },
    {
        "x": 123,
        "rtn": 321
    },
    {
        "x": 123,
        "rtn": 321
    },
    {
        "x": 123,
        "rtn": 321
    }]

s = Solution()
tester.uniTest(s.reverse, test_data)
````

**描述**：

请你来实现一个 myAtoi(string s) 函数，使其能将字符串转换成一个 32 位有符号整数（类似 C/C++ 中的 atoi 函数）。

函数 myAtoi(string s) 的算法如下：

+ 读入字符串并丢弃无用的前导空格
+ 检查下一个字符（假设还未到字符末尾）为正还是负号，读取该字符（如果有）。 确定最终结果是负数还是正数。 如果两者都不存在，则假定结果为正。
+ 读入下一个字符，直到到达下一个非数字字符或到达输入的结尾。字符串的其余部分将被忽略。
+ 将前面步骤读入的这些数字转换为整数（即，"123" -> 123， "0032" -> 32）。如果没有读入数字，则整数为 0 。必要时更改符号（从步骤 2 开始）。
+ 如果整数数超过 32 位有符号整数范围 [$−2^{31}$, $2^{31} − 1$] ，需要截断这个整数，使其保持在这个范围内。具体来说，小于 $−2^{31}$ 的整数应该被固定为 $−2^{31}$ ，大于 $2^{31} − 1$ 的整数应该被固定为 $2^{31} − 1$ 。
+ 返回整数作为最终结果。

**注意**：

+ 本题中的空白字符只包括空格字符 ' ' 。
+ 除前导空格或数字后的其余字符串外，**请勿忽略**任何其他字符。

**示例 1**：

```python
输入：s = "42"
输出：42
解释：加粗的字符串为已经读入的字符，插入符号是当前读取的字符。
第 1 步："42"（当前没有读入字符，因为没有前导空格）
         ^
第 2 步："42"（当前没有读入字符，因为这里不存在 '-' 或者 '+'）
         ^
第 3 步："42"（读入 "42"）
           ^
解析得到整数 42 。
由于 "42" 在范围 [-2^{31}, 2^{31} - 1] 内，最终结果为 42 。
```

**示例 2**：

```python
输入：s = "   -42"
输出：-42
解释：
第 1 步："   -42"（读入前导空格，但忽视掉）
            ^
第 2 步："   -42"（读入 '-' 字符，所以结果应该是负数）
             ^
第 3 步："   -42"（读入 "42"）

```

输入：s = "-91283472332"
输出：-2147483648

         ^

          ^
第 3 步："-91283472332"（读入 "91283472332"）
                     ^
解析得到整数 -91283472332 。
由于 -91283472332 小于范围 [-2^{31}, 2^{31} - 1] 的下界，最终结果被截断为 -2^{31} = -2147483648 。

+ s 由英文字母（大写和小写）、数字（0-9）、' '、'+'、'-' 和 '.' 组成

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["Traverse", "stateTransition"]


class Solution(object):
    def myAtoi(self, s: str) -> int:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(s)
        return rtn

    def Traverse(self, s: str) -> int:
        res = 0
        neg = False
        if not s:
            return res
        while s and s[0] == ' ':
            s = s[1:]
        if not s:
            return res
        if s[0] == '-':
            neg = True
            s = s[1:]
        elif s[0] == '+':
            s = s[1:]
        i = 0
        while i < len(s) and s[i] in "0123456789":
            i += 1
        s = s[:i]
        if not s:
            return res
        while s and s[0] == '0':
            s = s[1:]
        if not s:
            return res
        while s:
            res = res * 10 + int(s[0])
            s = s[1:]
        if neg and -res <= -2**31:
            return -2**31
        if not neg and res >= 2**31 - 1:
            return 2**31 - 1
        return -res if neg else res

    def stateTransition(self, s: str) -> int:
        states = [
            {' ': 0, 'd': 2, 's': 1, 'x': 3},  # 0
            {'d': 2, ' ': 3, 's': 3, 'x': 3},  # 1
            {'d': 2, ' ': 3, 's': 3, 'x': 3},  # 2
            {'d': 3, ' ': 3, 's': 3, 'x': 3},  # 3
        ]

        flag, ans, cur = 1, 0, 0
        for i in s:
            if i == ' ':
                t = ' '
            elif i in '+-':
                t = 's'
            elif i.isdigit():
                t = 'd'
            else:
                t = 'x'
            cur = states[cur][t]
            if cur == 2:
                ans = ans * 10 + int(i)
                ans = min(INT_MAX, ans) if flag == 1 else min(-INT_MIN, ans)
            elif cur == 1:
                flag = 1 if i == '+' else -1
        return flag * ans


test_data = [
    {
        "s": "42",
        "rtn": 42
    },
    {
        "s": "   -42",
        "rtn": -42
    },
    {
        "s": "4193 with words",
        "rtn": 4193
    },
    {
        "s": "words and 987",
        "rtn": 0
    },
    {
        "s": "-91283472332",
        "rtn": -2147483648
    },
    {
        "s": "00000-42a1234",
        "rtn": 0
    },
    {
        "s": "-2147483648",
        "rtn": -2147483648
    }]

s = Solution()
tester.uniTest(s.myAtoi, test_data)
````

# 9.回文数

**描述**：

给你一个整数 x ，如果 x 是一个回文整数，返回 true ；否则，返回 false 。

回文数是指正序（从左向右）和倒序（从右向左）读都是一样的整数。例如，121 是回文，而 123 不是。

**示例 1**：

```python
输入：x = 121
输出：true
```

**示例 2**：

```python
输入：x = -121
输出：false
解释：从左向右读, 为 -121 。 从右向左读, 为 121- 。因此它不是一个回文数。
```

**示例 3**：

```python
输入：x = 10
输出：false
解释：从右向左读, 为 01 。因此它不是一个回文数。
```

**示例 4**：

```python
输入：x = -101
输出：false
```

**提示**：

+ $-2^{31} <= x <= 2^{31} - 1$

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["toString", "reverse", "halfReverse"]


class Solution(object):
    def isPalindrome(self, x: int) -> bool:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(x)
        return rtn

    def toString(self, x: int) -> bool:
        if x < 0:
            return False
        x = str(x)
        while not (len(x) == 0 or len(x) == 1):
            if x[0] != x[-1]:
                return False
            else:
                x = x[1:-1]
        return True

    def reverse(self, x: int) -> bool:
        return x >= 0 and int(str(x)[::-1]) == x

    def halfReverse(self, x: int) -> bool:
        """
        只反转后面一半的数字!!(节省一半的时间)
        """
        if x < 0 or (x != 0 and x % 10 == 0):
            return False
        elif x == 0:
            return True
        reverse_x = 0
        while x > reverse_x:
            remainder = x % 10
            reverse_x = reverse_x * 10 + remainder
            x = x // 10
        # 当x为奇数时, 只要满足 reverse_x//10 == x 即可
        if reverse_x == x or reverse_x // 10 == x:
            return True
        return False


test_data = [
    {
        "x": 121,
        "rtn": True
    },
    {
        "x": -121,
        "rtn": False
    },
    {
        "x": 10,
        "rtn": False
    },
    {
        "x": -101,
        "rtn": False
    }]

s = Solution()
tester.uniTest(s.isPalindrome, test_data)
````

# 10.正则表达式匹配

**描述**：

给你一个字符串 `s` 和一个字符规律 `p`，请你来实现一个支持 `'.'` 和 `'*'` 的正则表达式匹配。

+ '.' 匹配任意单个字符
+ '*' 匹配零个或多个前面的那一个元素

所谓匹配，是要涵盖 **整个** 字符串 `s` 的，而不是部分字符串。

**示例 1**：

```python
输入：s = "aa" p = "a"
输出：false
解释："a" 无法匹配 "aa" 整个字符串。
```

**示例 2**：

```python
输入：s = "aa" p = "a*"
输出：true
解释：因为 '*' 代表可以匹配零个或多个前面的那一个元素, 在这里前面的元素就是 'a'。因此，字符串 "aa" 可被视为 'a' 重复了一次。
```

**示例 3**：

```python
输入：s = "ab" p = ".*"
输出：true
解释：".*" 表示可匹配零个或多个（'*'）任意字符（'.'）。
```

**示例 4**：

```python
输入：s = "aab" p = "c*a*b"
输出：true
解释：因为 '*' 表示零个或多个，这里 'c' 为 0 个, 'a' 被重复一次。因此可以匹配字符串 "aab"。
```

**示例 5**：

```python
输入：s = "mississippi" p = "mis*is*p*."
输出：false
```

**提示**：

+ 1 <= s.length <= 20
+ 1 <= p.length <= 30
+ s 只含小写英文字母。
+ p 只含小写英文字母，以及字符 . 和 *。
+ 保证每次出现字符 * 时，前面都匹配到有效的字符

````python
# -*- encoding:utf-8 -*-

from utils import tester

# starSplit is not completed
SOLUTE = ["starSplit", "dp"]


class Solution(object):
    def isMatch(self, s: str, p: str) -> bool:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(s, p)
        return rtn

    def starSplit(self, s: str, p: str) -> bool:
        tmp = chr(ord(max(p))+1)
        plist = p.replace("*", "*" + tmp).split(tmp)
        if not plist[-1]:
            plist.pop(-1)
        print("raw plist", plist)
        if plist[-1][-1] != "*":
            for c in plist[-1][::-1]:
                if s and (s[-1] == c or c == "."):
                    s = s[:-1]
                else:
                    return False
            plist.pop(-1)
        for i in range(len(plist) - 1):
            c = plist[i][-2]
            if c != ".":
                while plist[i + 1][:-2] and plist[i + 1][0] == c:
                    plist[i + 1] = plist[i + 1][1:]
            if plist[i] == plist[i + 1]:
                plist[i] = ""
        plist = [ps for ps in plist if ps]
        print(plist, s)
        flag = False
        for i, ps in enumerate(plist):
            if not s:
                if len(ps) == 2:
                    continue
                return False
            if ps[-1] == "*":
                if ps[:-2]:
                    for c in ps[:-2]:
                        if not s:
                            return False
                        while s:
                            if s[0] == c or c == ".":
                                flag = False
                                s = s[1:]
                                break
                            elif flag:
                                s = s[1:]
                            else:
                                return False
                if ps[-2] == ".":
                    flag = True
                    print(ps, s)
                    continue
                else:
                    while s:
                        if s[0] == ps[-2]:
                            flag = False
                            s = s[1:]
                        elif flag:
                            s = s[1:]
                        else:
                            break
            print(ps, s)
        if s:
            return True if flag else False
        return True

    def dp(self, s: str, p: str) -> bool:
        def match(charS: str, charP: str) -> bool:
            '''当正则字符为'.'，或二者字符相等时，返回True'''
            return charP == '.' or charP == charS

        # dp[i][j]表示 [正则串p的前i个字符] 能否配对成功 [目标串s的前j个字符]
        # 为了更多的cache hit，我们将逐个遍历i，即逐个读入p串的字符
        lenS, lenP = len(s), len(p)
        dp = [[False] * (lenS + 1) for _ in range(lenP + 1)]

        # 空字符配对空字符则必定成功
        dp[0][0] = True

        # 如果正则串为‘X*X*X*...’的形式，则其可配对空字符串
        for i in range(1, lenP + 1):
            if p[i - 1] == '*':
                dp[i][0] = dp[i - 2][0]

        for i in range(1, lenP + 1):
            for j in range(1, lenS + 1):
                # 如果p新读入的字符是'*'，则需要比对'*'之前的字符与s的相应字符
                if p[i - 1] == '*':
                    # 1. 如果两个字符能够配对，我们用'X'表示这个字符，有两种情况：
                    # 1.1. 不使用'X*'，即p串除去'X*'的部分可以与s串配对，即dp[i-2][j]
                    # 1.2. 使用'X*'，因为我们已经确定'X'能够与s串的字符配对，因此，
                    #      只要p串之前的某一个字符与s串的这个字符配对成功过，之后'*'这行所有
                    #      能够配对的字符都能配对成功，dp[i][j-1]
                    #      e.g. 'ab*' 只要与'abbbb...'在第三行第一列配对成功了，即'ab*'与'a'的dp[3][1]
                    #           之后对于所有'bbbbb...'，dp[3][...]都是True
                    if match(s[j-1], p[i-2]):
                        dp[i][j] = dp[i-2][j] or dp[i][j-1]

                    # 2. 如果两个字符配对不成功
                    # 2.1. 此时只有不使用'X*'一种方法能够使p,s配对，即dp[i-2][j]
                    else:
                        dp[i][j] = dp[i-2][j]

                # 如果p新读入的不是'*'，就很简单了，只要[配对]且[二者上一个字符能够配对]就可以
                else:
                    dp[i][j] = match(s[j-1], p[i-1]) and dp[i-1][j-1]

        return dp[-1][-1]


test_data = [
    {
        "s": "baabbbaccbccacacc",
        "p": "c*..b*a*a.*a..*c",
        "rtn": True
    },
    {
        "s": "aabcbcbcaccbcaabc",
        "p": ".*a*aa*.*b*.c*.*a*",
        "rtn": True
    },
    {
        "s": "a",
        "p": ".*..a*",
        "rtn": False
    },
    {
        "s": "aab",
        "p": "c*a*b",
        "rtn": True
    },
    {
        "s": "mississippi",
        "p": "mis*is*p*.",
        "rtn": False
    },
    {
        "s": "ab",
        "p": ".*c",
        "rtn": False
    }]

s = Solution()
tester.uniTest(s.isMatch, test_data)
````

# 11.盛最多水的容器

**描述**：

给你 n 个非负整数 a1，a2，...，an，每个数代表坐标中的一个点 (i, ai) 。在坐标内画 n 条垂直线，垂直线 i 的两个端点分别为 (i, ai) 和 (i, 0) 。找出其中的两条线，使得它们与 x 轴共同构成的容器可以容纳最多的水。

**说明**：

你不能倾斜容器。

**示例 1**：

<center>
<img src="https://aliyun-lc-upload.oss-cn-hangzhou.aliyuncs.com/aliyun-lc-upload/uploads/2018/07/25/question_11.jpg"/>
</center>

```python
输入：[1,8,6,2,5,4,8,3,7]
输出：49
解释：图中垂直线代表输入数组 [1,8,6,2,5,4,8,3,7]。在此情况下，容器能够容纳水（表示为蓝色部分）的最大值为 49。
```

**示例 2**：

```python
输入：height = [1,1]
输出：1
```

**示例 3**：

```python
输入：height = [4,3,2,1,4]
输出：16
```

**示例 4**：

```python
输入：height = [1,2,1]
输出：2
```

**提示**：

+ n == height.length
+ 2 <= n <= $10^{5}$
+ 0 <= height[i] <= $10^{4}$

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["Traverse", "twoPointer"]


class Solution(object):
    def maxArea(self, height: List[int]) -> int:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(height)
        return rtn

    def Traverse(self, height: List[int]) -> int:
        N = len(height)
        maxVolume = 0

        for i in range(N - 1):
            for j in range(i, N):
                tmpV = min(height[i], height[j]) * (j - i)
                if maxVolume < tmpV:
                    maxVolume = tmpV

        return maxVolume

    def twoPointer(self, height: List[int]) -> int:
        maxh = max(height)  # 最高的点/柱子
        i, j, res = 0, len(height) - 1, 0
        while i < j:
            if res > maxh * (j - i):
                break
            if height[i] < height[j]:
                res = max(res, height[i] * (j - i))
                i += 1
            else:
                res = max(res, height[j] * (j - i))
                j -= 1
        return res


test_data = [
    {
        "height": [1, 8, 6, 2, 5, 4, 8, 3, 7],
        "rtn": 49
    },
    {
        "height": [1, 1],
        "rtn": 1
    },
    {
        "height": [4, 3, 2, 1, 4],
        "rtn": 16
    },
    {

        "height": [1, 2, 1],
        "rtn": 2
    }]

s = Solution()
tester.uniTest(s.maxArea, test_data)
````

# 12.整数转罗马数字

**描述**：

罗马数字包含以下七种字符： I， V， X， L，C，D 和 M。

```
字符          数值
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
```

例如， 罗马数字 2 写做 II ，即为两个并列的 1。12 写做 XII ，即为 X + II 。 27 写做  XXVII, 即为 XX + V + II 。

通常情况下，罗马数字中小的数字在大的数字的右边。但也存在特例，例如 4 不写做 IIII，而是 IV。数字 1 在数字 5 的左边，所表示的数等于大数 5 减小数 1 得到的数值 4 。同样地，数字 9 表示为 IX。这个特殊的规则只适用于以下六种情况：

+ I 可以放在 V (5) 和 X (10) 的左边，来表示 4 和 9。
+ X 可以放在 L (50) 和 C (100) 的左边，来表示 40 和 90。 
+ C 可以放在 D (500) 和 M (1000) 的左边，来表示 400 和 900。

给你一个整数，将其转为罗马数字。

**示例 1**：

```python
输入: num = 3
输出: "III"
```

**示例 2**：

```python
输入: num = 4
输出: "IV"
```

**示例 3**：

```python
输入: num = 9
输出: "IX"
```

**示例 4**：

```python
输入: num = 58
输出: "LVIII"
解释: L = 50, V = 5, III = 3.
```

**示例 5**：

```python
输入: num = 1994
输出: "MCMXCIV"
解释: M = 1000, CM = 900, XC = 90, IV = 4.
```

**提示**：

+ 1 <= num <= 3999

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["Traverse", "hashMap"]


class Solution(object):
    def intToRoman(self, num: int) -> str:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(num)
        return rtn

    def Traverse(self, num: int) -> str:
        Roman_dict = {
            1: "I",
            5: "V",
            10: "X",
            50: "L",
            100: "C",
            500: "D",
            1000: "M"
        }
        # {0: 1, 1: 5, 2: 10, 3: 50, 4: 100, 5: 500, 6: 1000}
        Roman_id = dict(zip(range(len(Roman_dict)), Roman_dict.keys()))

        import math

        def recursive(n: int, res: str) -> str:
            if n == 0:
                return res
            l = int(math.log10(n))
            tmp = int(n / (10 ** l))
            if tmp <= 3:
                res += tmp * Roman_dict[Roman_id[2 * l]]
            elif tmp == 4:
                res += Roman_dict[Roman_id[2 * l]] + \
                    Roman_dict[Roman_id[2 * l + 1]]
            elif tmp == 5:
                res += Roman_dict[Roman_id[2 * l + 1]]
            elif tmp >= 6 and tmp <= 8:
                res += Roman_dict[Roman_id[2 * l + 1]] + \
                    (tmp - 5) * Roman_dict[Roman_id[2 * l]]
            elif tmp == 9:
                res += Roman_dict[Roman_id[2 * l]] + \
                    Roman_dict[Roman_id[2 * l + 2]]

            if l == 0:
                return res
            else:
                res = recursive(int(n % (10 ** l)), res)
                return res

        result = ""
        result = recursive(num, result)

        return result

    def hashMap(self, num: int) -> str:
        # 使用哈希表，按照从大到小顺序排列
        hashmap = {1000: 'M', 900: 'CM', 500: 'D', 400: 'CD', 100: 'C',
                   90: 'XC', 50: 'L', 40: 'XL', 10: 'X', 9: 'IX', 5: 'V', 4: 'IV', 1: 'I'}
        res = ''
        for key in hashmap:
            if num // key != 0:
                count = num // key  # 比如输入4000，count 为 4
                res += hashmap[key] * count
                num %= key
        return res


test_data = [
    {
        "num": 10,
        "rtn": "X"
    },
    {
        "num": 3,
        "rtn": "III"
    },
    {
        "num": 4,
        "rtn": "IV"
    },
    {
        "num": 9,
        "rtn": "IX"
    },
    {
        "num": 58,
        "rtn": "LVIII"
    },
    {
        "num": 1994,
        "rtn": "MCMXCIV"
    }]

s = Solution()
tester.uniTest(s.intToRoman, test_data)
````

# 13.罗马数字转整数

**描述**：

罗马数字包含以下七种字符： I， V， X， L，C，D 和 M。

```
字符          数值
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
```

例如， 罗马数字 2 写做 II ，即为两个并列的 1。12 写做 XII ，即为 X + II 。 27 写做  XXVII, 即为 XX + V + II 。

通常情况下，罗马数字中小的数字在大的数字的右边。但也存在特例，例如 4 不写做 IIII，而是 IV。数字 1 在数字 5 的左边，所表示的数等于大数 5 减小数 1 得到的数值 4 。同样地，数字 9 表示为 IX。这个特殊的规则只适用于以下六种情况：

+ I 可以放在 V (5) 和 X (10) 的左边，来表示 4 和 9。
+ X 可以放在 L (50) 和 C (100) 的左边，来表示 40 和 90。 
+ C 可以放在 D (500) 和 M (1000) 的左边，来表示 400 和 900。

给定一个罗马数字，将其转换成整数。

**示例 1**：

```python
输入: s = "III"
输出: 3
```

**示例 2**：

```python
输入: s = "IV"
输出: 4
```

**示例 3**：

```python
输入: s = "IX"
输出: 9
```

**示例 4**：

```python
输入: s = "LVIII"
输出: 58
解释: L = 50, V= 5, III = 3.
```

**示例 5**：

```python
输入: s = "MCMXCIV"
输出: 1994
解释: M = 1000, CM = 900, XC = 90, IV = 4.
```

**提示**：

+ 1 <= s.length <= 15
+ s 仅含字符 ('I', 'V', 'X', 'L', 'C', 'D', 'M')
+ 题目数据保证 s 是一个有效的罗马数字，且表示整数在范围 [1, 3999] 内
+ 题目所给测试用例皆符合罗马数字书写规则，不会出现跨位等情况。
+ IL 和 IM 这样的例子并不符合题目要求，49 应该写作 XLIX，999 应该写作 CMXCIX 。
+ 关于罗马数字的详尽书写规则，可以参考 [罗马数字 - Mathematics](https://b2b.partcommunity.com/community/knowledge/zh_CN/detail/10753/%E7%BD%97%E9%A9%AC%E6%95%B0%E5%AD%97#knowledge_article) 。

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["hashMap"]


class Solution(object):
    def romanToInt(self, s: str) -> int:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(s)
        return rtn

    def hashMap(self, s: str) -> int:
        d = {'I': 1, 'IV': 4, 'V': 5, 'IX': 9, 'X': 10, 'XL': 40, 'L': 50,
             'XC': 90, 'C': 100, 'CD': 400, 'D': 500, 'CM': 900, 'M': 1000}
        result = 0
        i = 0
        while i < len(s):
            # 查看当前位和下一位的字符
            str1 = s[i:i+2]
            # 如果当前位置是特殊情况，那么返回其在字典中对应值，并且下一次从特殊字符之后一位开始索引
            if str1 in d:
                result += d[str1]
                i += 2
            # 如果当前位不是特殊情况，那么只返回当前位的数值
            else:
                result += d[s[i]]
                i += 1
        return result


test_data = [
    {
        "s": "X",
        "rtn": 10
    },
    {
        "s": "III",
        "rtn": 3
    },
    {
        "s": "IV",
        "rtn": 4
    },
    {
        "s": "IX",
        "rtn": 9
    },
    {
        "s": "LVIII",
        "rtn": 58
    },
    {
        "s": "MCMXCIV",
        "rtn": 1994
    }]

s = Solution()
tester.uniTest(s.romanToInt, test_data)
````

# 14.最长公共前缀

**描述**：

编写一个函数来查找字符串数组中的最长公共前缀。

如果不存在公共前缀，返回空字符串 ""。

**示例 1**：

```python
输入：strs = ["flower","flow","flight"]
输出："fl"
```

**示例 2**：

```python
输入：strs = ["dog","racecar","car"]
输出：""
解释：输入不存在公共前缀。
```

**提示**：

+ 1 <= strs.length <= 200
+ 0 <= strs[i].length <= 200
+ strs[i] 仅由小写英文字母组成

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["Traverse", "Zip", "minmax"]


class Solution(object):
    def longestCommonPrefix(self, strs: List[str]) -> str:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(strs)
        return rtn

    def Traverse(self, strs: List[str]) -> str:
        N = len(strs)
        if N == 1:
            return strs[0]
        res = strs.pop()
        if not res:
            return ""
        for i in range(N - 1):
            tmp = strs.pop()
            if not tmp:
                return ""
            n = min(len(tmp), len(res))
            flag = False
            for j in range(n):
                if tmp[j] == res[j]:
                    continue
                else:
                    flag = True
                    break
            if not flag:
                res = res[:n]
            else:
                res = res[:j]
            if not res:
                break
        return res

    def Zip(self, strs: List[str]) -> str:
        ans = ""
        for i in list(zip(*strs)):
            if len(set(i)) == 1:
                ans += i[0]
            else:
                break
        return ans

    def minmax(self, strs: List[str]) -> str:
        if not strs:
            return ""
        str0 = min(strs)
        str1 = max(strs)
        for i in range(len(str0)):
            if str0[i] != str1[i]:
                return str0[:i]
        return str0


test_data = [
    {
        "strs": ["flower", "flow", "flight"],
        "rtn": "fl"
    },
    {
        "strs": ["dog", "racecar", "car"],
        "rtn": ""
    }]

s = Solution()
tester.uniTest(s.longestCommonPrefix, test_data)
````

# 15.三数之和

**描述**：

给你一个包含 n 个整数的数组 nums，判断 nums 中是否存在三个元素 a，b，c ，使得 a + b + c = 0 ？请你找出所有和为 0 且不重复的三元组。

**注意**：

答案中不可以包含重复的三元组。

**示例 1**：

```python
输入：nums = [-1,0,1,2,-1,-4]
输出：[[-1,-1,2],[-1,0,1]]
```

**示例 2**：

```python
输入：nums = []
输出：[]
```

**示例 3**：

```python
输入：nums = [0]
输出：[]
```

**提示**：

+ 0 <= nums.length <= 3000
+ $-10^{5}$ <= nums[i] <= $10^{5}$

````python
# -*- encoding:utf-8 -*-

from utils import tester

SOLUTE = ["twoPointer", ]


class Solution(object):
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        Solute = getattr(self, SOLUTE[-1])
        rtn = Solute(nums)
        return rtn

    def twoPointer(self, nums: List[int]) -> List[List[int]]:
        if len(nums) < 3:
            return []
        nums.sort()
        n = len(nums)
        res = []
        for i in range(n):
            if(nums[i] > 0):
                return res
            if(i > 0 and nums[i] == nums[i - 1]):
                continue
            L = i + 1
            R = n - 1
            while(L < R):
                if(nums[i] + nums[L] + nums[R] == 0):
                    res.append([nums[i], nums[L], nums[R]])
                    while(L < R and nums[L] == nums[L + 1]):
                        L = L + 1
                    while(L < R and nums[R] == nums[R - 1]):
                        R = R - 1
                    L = L + 1
                    R = R - 1
                elif(nums[i] + nums[L] + nums[R] > 0):
                    R = R - 1
                else:
                    L = L + 1
        return res


test_data = [
    {
        "nums": [-1, 0, 1, 2, -1, -4],
        "rtn": [[-1, -1, 2], [-1, 0, 1]]
    },
    {
        "nums": [],
        "rtn": []
    },
    {
        "nums": [0],
        "rtn": []
    }]

s = Solution()
tester.uniTest(s.threeSum, test_data)
````
